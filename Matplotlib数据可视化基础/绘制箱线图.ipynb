```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文及字符显示
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

# 读取数据
data = np.load('data/国民经济核算季度数据.npz',allow_pickle=True)
name = data['columns']
values = data['values']
print(name)
print(values)
```

```python
label = ['第一产业','第二产业','第三产业']

# 绘制画布
plt.figure(figsize=(6,4))

#绘制箱线图
plt.boxplot(values[:,3:6],notch=True,labels=label,meanline=True)
plt.title("2000-2017年各产业国民生产总值箱线图")
plt.savefig('tmpsave/2000-2017年各产业国民生产总值箱线图.png')
plt.show()
```

```python
label = ['第一产业','第二产业','第三产业']
gdb = (list(values[:,3]),list(values[:,4]),list(values[:,5]))

# 绘制画布
plt.figure(figsize=(6,4))

#绘制箱线图
plt.boxplot(gdb,notch=True,labels=label,meanline=True)
plt.title("2000-2017年各产业国民生产总值箱线图")
plt.savefig('tmpsave/2000-2017年各产业国民生产总值箱线图.png')
plt.show()
```
