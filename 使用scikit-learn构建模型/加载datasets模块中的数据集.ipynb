```python
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_iris
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits
from sklearn.datasets import load_boston
from sklearn.datasets import fetch_california_housing

cancer = load_breast_cancer() # 将数据集赋值给cancer变量
print('breast_cancer数据集的长度为：\n',len(cancer))
print('breast_cancer数据集的类型为：\n',type(cancer))

iris = load_iris() # 将数据集赋值给iris变量
print('iris数据集的长度为：\n',len(iris))
print('iris数据集的类型为：\n',type(iris))

wine = load_wine() # 将数据集赋值给wine变量
print('wine数据集的长度为：\n',len(wine))
print('wine数据集的类型为：\n',type(wine))

digits = load_digits() # 将数据集赋值给digits变量
print('digits数据集的长度为：\n',len(digits))
print('digits数据集的类型为：\n',type(digits))

boston = load_boston() # 将数据集赋值给boston变量
print('boston数据集的长度为：\n',len(boston))
print('boston数据集的类型为：\n',type(boston))

fetch_california_housing = fetch_california_housing() # 将数据集赋值给fetch_california_housing变量
print('fetch_california_housing数据集的长度为：\n',len(fetch_california_housing))
print('fetch_california_housing数据集的类型为：\n',type(fetch_california_housing))
```

```python
cancer_data = cancer['data']
print('breast_cancer数据集的数据为：\n',cancer_data)
iris_data = iris['data']
print('iris数据集的数据为：\n',iris_data)
wine_data = wine['data']
print('wine数据集的数据为：\n',wine_data)
digits_data = digits['data']
print('digits数据集的数据为：\n',digits_data)
boston_data = boston['data']
print('boston数据集的数据为：\n',boston_data)
fetch_california_housing_data = fetch_california_housing['data']
print('fetch_california_housing数据集的数据为：\n',fetch_california_housing_data)
```

```python
cancer_target = cancer['target'] # 取出数据集的标签
print('breast_cancer数据集的标签为：\n',cancer_target)
iris_target = iris['target'] # 取出数据集的标签
print('iris数据集的标签为：\n',iris_target)
wine_target = wine['target'] # 取出数据集的标签
print('wine数据集的标签为：\n',wine_target)
digits_target = digits['target'] # 取出数据集的标签
print('digits数据集的标签为：\n',digits_target)
boston_target = boston['target'] # 取出数据集的标签
print('boston数据集的标签为：\n',boston_target)
fetch_california_housing_target = fetch_california_housing['target'] # 取出数据集的标签
print('fetch_california_housing数据集的标签为：\n',fetch_california_housing_target)
```

```python
cancer_names = cancer['feature_names'] # 取出数据集的特征名
print('breast_names数据集的特征名为：\n',cancer_names)
iris_names = iris['feature_names']  # 取出数据集的特征名
print('iris数据集的特征名为：\n',iris_names)
wine_names = wine['feature_names']  # 取出数据集的特征名
print('wine数据集的特征名为：\n',wine_names)
# digits_names = digits['feature_names']  # 取出数据集的特征名
# print('digits数据集的特征名为：\n',digits_names)
boston_names = boston['feature_names']  # 取出数据集的特征名
print('boston数据集的特征名为：\n',boston_names)
fetch_california_housing_names = fetch_california_housing['feature_names']  # 取出数据集的特征名
print('fetch_california_housing数据集的特征名为：\n',fetch_california_housing_names)
```

```python
cancer_desc = cancer['DESCR'] # 取出数据集的描述信息
print('breast_names数据集的描述信息为：\n',cancer_desc )
print('*'*100)
iris_desc  = iris['DESCR']  # 取出数据集的描述信息
print('iris数据集的描述信息为：\n',iris_desc )
print('*'*100)
wine_desc  = wine['DESCR']  # 取出数据集的描述信息
print('wine数据集的描述信息为：\n',wine_desc )
print('*'*100)
digits_desc  = digits['DESCR'] # 取出数据集的描述信息
print('digits数据集的描述信息为：\n',digits_desc )
print('*'*100)
boston_desc  = boston['DESCR']  # 取出数据集的描述信息
print('boston数据集的描述信息为：\n',boston_desc )
print('*'*100)
fetch_california_housing_desc  = fetch_california_housing['DESCR']  # 取出数据集的描述信息
print('fetch_california_housing数据集的描述信息为：\n',fetch_california_housing_desc )
```

```python
print('原始数据集数据的形状为：',cancer_data.shape)
print('原始数据集标签的形状为：',cancer_target.shape)
```

```python
from sklearn.model_selection import train_test_split

cancer_data_train,cancer_data_test,cancer_target_train,cancer_target_test = train_test_split(cancer_data,cancer_target,
                                                                                             test_size=0.2,random_state=42)
print('训练集数据的形状为：',cancer_data_train.shape)
print('训练集数据的标签形状为：',cancer_target_train.shape)
print('测试集数据的形状为：',cancer_data_test.shape)
print('测试集数据的标签形状为：',cancer_target_test.shape)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 生成规则
Scaler = MinMaxScaler().fit(cancer_data_train)
# 将规则应用于训练集
cancer_trainScaler = Scaler.transform(cancer_data_train)
# 将规则应用于测试集
cancer_testScaler = Scaler.transform(cancer_data_test)
print('离差标准化前训练集数据的最小值：',cancer_data_train.min())
print('离差标准化后训练集数据的最小值：',np.min(cancer_trainScaler))
print('离差标准化前训练集数据的最大值：',np.max(cancer_data_train))
print('离差标准化后训练集数据的最大值：',np.max(cancer_trainScaler))
print('离差标准化前测试集数据的最小值：',np.min(cancer_data_test))
print('离差标准化后测试集数据的最小值：',np.min(cancer_testScaler))
print('离差标准化前测试集数据的最大值：',np.max(cancer_data_test))
print('离差标准化后测试集数据的最大值：',np.max(cancer_testScaler))
```

```python
from sklearn.decomposition import PCA
 
# 生成规则
pca_model=PCA(n_components=10).fit(cancer_trainScaler)
# 将规则应用到训练集
cancer_trainPca = pca_model.transform(cancer_trainScaler)
# 将规则应用到测试集
cancer_testPca = pca_model.transform(cancer_testScaler)
print('PCA降维前训练集数据的形状为：',cancer_trainScaler.shape)
print('PCA降维后训练集数据的形状为：',cancer_trainPca.shape)
print('PCA降维前测试集数据的形状为：',cancer_testScaler.shape)
print('PCA降维后测试集数据的形状为：',cancer_testPca.shape)
```
